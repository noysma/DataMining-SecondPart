# PCA

```python
pca = PCA(n_components=100, whiten=True).fit(X_train)

plt.rcParams["figure.figsize"] = (12,7)

fig, ax = plt.subplots()
xi = np.arange(1, 101, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.5,1.1)
plt.plot(xi, y, marker='o', linestyle='-', color='black')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 105, step=5), rotation=60)
plt.ylabel('Cumulative variance (%)')
plt.title('Number of components needed to explain variance \n', weight='bold')

plt.axhline(y=0.95, color='red', linestyle='--')
plt.text(1, 0.98, '95% cut-off threshold', color = 'black', fontsize=16)

ax.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
pca = PCA(n_components=65)
X_train_pca = pca.fit_transform(X_train)
```

```python
X_train_pca.shape
```

```python
plt.figure(figsize=(20, 13))
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 2], c=y_train, cmap='prism', edgecolor='k', alpha=0.7)
plt.show()
#da mettere (0,1) o (0,2)
```

```python
fig = plt.figure(figsize=(20, 13))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train_pca[:, 0], X_train_pca[:, 1], X_train_pca[:, 2], cmap='prism', edgecolor='k')
ax.view_init(4, -72)
```

# DT PCA

```python
X_test_pca = pca.transform(X_test)

clf = DecisionTreeClassifier(min_samples_leaf=3, random_state=42)
clf.fit(X_train_pca, y_train)

y_pred = clf.predict(X_test_pca)

print(classification_report(y_test, y_pred))
```

```python
y_score = clf.predict_proba(X_test_pca)
```

```python
plot_roc(y_test, y_score, figsize=(13,10))
plt.show()
```
