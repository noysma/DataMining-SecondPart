```python
#Prepare initial centers - amount of initial center defines amount of clusters from which X-Means will
# start analysis.
amount_initial_centers = 2
initial_centers = kmeans_plusplus_initializer(X_train.values,amount_initial_centers).initialize()
# Create instance of X-Means algorithm. The algorithm will start analysis from 2 clusters, the maximum
# number of cluster that can be allocated is 20.
clusters = []
centers = []
error = []
for i in range(100):
    xmeans_instance = xmeans (X_train.values, initial_centers, i)
    xmeans_instance.process()
    # Extract clustering results: clusters and their centers
    #clusters.append(xmeans_instance.get_clusters())
    #centers.append(xmeans_instance.get_centers())
    # Print total sum of metric errors
    error.append(xmeans_instance.get_total_wce())
    #print ("Total WCE:", xmeans_instance.get_total_wce())
```

```python
plt.rcParams['axes.facecolor'] = 'w'
plt.plot(np.arange(100), error, markerfacecolor='black')
plt.ylabel('WCE score \n', fontsize = 15, fontweight = 'bold')
plt.xlabel('\n Values of K', fontsize = 15, fontweight = 'bold')
plt.xticks(np.arange(0,106,15))
plt.yticks(rotation=55)
plt.grid(True)
plt.show()
```

```python
def plot_two_axes(fig, x1,y1,x2,y2,ax1,labels=['A','B'],colors=['red','blue']):
    color=colors[0]
    ax1.set_xlabel('')
    ax1.set_ylabel(labels[0], color=color)
    ax1.plot(x1 ,y1, color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
    color=colors[1]
    ax2.set_ylabel(labels[1], color=color)  # we already handled the x-label with ax1
    ax2.plot(x2,y2, color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    fig.tight_layout()  # otherwise the right y-label is slightly clippedd

fig, ax =plt.subplots(1,2,figsize=(25,10))
plot_two_axes(fig, range(2, len(sse_list)+2), sse_list, range_n_clusters, silhouette_avg, ax[0],labels=['SSE','Silhouette'])
plt.grid(True)
plt.show()
```

# OPTICS

```python
from sklearn.cluster import OPTICS
```

```python
optics = OPTICS(min_samples=20, max_eps=np.inf, cluster_method='xi')
optics.fit(X_train)
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.gridspec as gridspec
```

```python
values = np.arange(2, 24, 3)
plt.rcParams['axes.facecolor'] = 'w'
range_n_samples = np.arange(2,24)
silhouette_avg = []
for n_samples in range_n_samples:
    optics = OPTICS(min_samples=n_samples, max_eps=np.inf, cluster_method='xi')
    optics.fit(X_train)
    cluster_labels = optics.labels_
    silhouette_avg.append(silhouette_score(X_train, optics.labels_))
```

```python
plt.plot(silhouette_avg, marker='.', markersize=10, markerfacecolor='black')
plt.xlabel('\n Values of K', fontsize = 15, fontweight = 'bold')
plt.ylabel('Silhouette score \n', fontsize = 15, fontweight = 'bold')
plt.xticks(values)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
G = gridspec.GridSpec(2, 3)
ax1 = plt.subplot(G[0, :])

space = np.arange(len(X_train))
reachability = optics.reachability_[optics.ordering_]
labels = optics.labels_[optics.ordering_]

colors = ["b.", "r.", "g.", "y.", "b."]
for klass, color in zip(range(0, 5), colors):
    Xk = space[labels == klass]
    Rk = reachability[labels == klass]
    ax1.plot(Xk, Rk, color, alpha=0.3)
ax1.plot(space[labels == -1], reachability[labels == -1], "b.", alpha=0.3)
ax1.plot(space, np.full_like(space, 2.0, dtype=float), "k-", alpha=0.5)
ax1.plot(space, np.full_like(space, 0.5, dtype=float), "k-.", alpha=0.5)
ax1.set_ylabel("Reachability (epsilon distance)")
ax1.set_title("Reachability Plot")

plt.show()
```

```python
clust_name = ['Cluster 2', 'Cluster 1']
colors = ['deeppink', 'dodgerblue']
for cluster_id in np.unique(optics.labels_)[:20]:
    indexes = np.where(optics.labels_== cluster_id)
    plt.scatter(X_train.values[indexes,3], X_train.values[indexes,4], label = clust_name[cluster_id], c = colors[cluster_id], alpha=0.4)
plt.legend(fontsize=10, loc='best')
plt.show()
```

# K-Mode

https://github.com/nicodv/kmodes

```python
from kmodes.kmodes import KModes
```

```python
X_trasposto = X_train.T
```

```python
X_cat = list()
for i in range(len(X_trasposto)):
    X_cat.append(pd.qcut(X_trasposto[i], q=[0.0, 0.25, 0.5, 0.75, 1.0], labels=False))
X_cat = np.array(X_cat)
```

```python
cost = []
for num_cluster in list(range(1,20)):
    kmode = KModes (n_clusters=num_cluster, init='Huang', n_init=1, verbose=1)
    kmode.fit(X_cat)
    cost.append(kmode.cost_)
```

```python
plt.rcParams['axes.facecolor'] = 'w'
plt.plot(np.arange(19), cost, markerfacecolor='black')
plt.ylabel('Cost Function \n', fontsize = 15, fontweight = 'bold')
plt.xlabel('\n Values of K', fontsize = 15, fontweight = 'bold')
plt.xticks(np.arange(0,22,3))
#plt.yticks(rotation=55)
plt.grid(True)
plt.show()
```

```python
km = KModes(n_clusters=5, init='Huang', n_init=5, verbose=1)

clusters = km.fit_predict(X_cat)
```

```python
centers = km.cluster_centroids_
```

```python
km.labels_
```

```python
clust_name = ['1', '2', '3', '4', '5']
for cluster_id in np.unique(km.labels_)[:20]:
    indexes = np.where(km.labels_== cluster_id)
    plt.scatter(X_train.values[indexes,3], X_train.values[indexes,4], alpha=0.4, label = clust_name[cluster_id])
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
clust_name = ['1', '2', '3', '4', '5']
colors = ['#A6CEE3', '#FDBF6F', '#B15928', '#58AA4D', '#ff66ff']
for i in range(len(centers)):
    plt.plot(centers[i], marker='o', label = clust_name[i], c=colors[i])
    plt.xlim(20, 60)


plt.yticks(fontsize=15)
plt.legend(fontsize=15, loc='best')
plt.grid(axis='y')
```
